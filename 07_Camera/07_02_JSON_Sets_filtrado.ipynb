```python
import pandas as pd
import json

# Cargar el archivo CSV
df_lego = pd.read_csv("../01_Data_Cleaning/df_lego_final.csv")

# Cargar el archivo JSON con los números de sets deseados
with open("../07_Camera/idx_to_class.json", "r", encoding="utf-8") as file:
    data = json.load(file)

# Extraer los números de set del JSON
set_numbers = list(data.values())

# Filtrar el DataFrame para que solo contenga los sets deseados
df_filtrado = df_lego[df_lego["Number"].astype(str).isin(set_numbers)]

# Seleccionar solo las columnas requeridas y eliminar las demás
df_filtrado = df_filtrado[["Number", "YearFrom", "Theme", "SetName", 
                                    "USRetailPrice", "WantCount", "BrickLinkSoldPriceUsed"]]

# Guardar el nuevo CSV filtrado
filtered_csv_path = "df_lego_camera.csv"
df_filtrado.to_csv(filtered_csv_path, index=False)

```
