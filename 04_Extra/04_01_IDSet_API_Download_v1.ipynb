***Código de redimensionadod e imágenes intentando tener la menos pérdida de información y mayor precisión para el ML***

```python
import cv2
import numpy as np
import os

# Tamaño de la imagen final
IMAGE_SIZE = (250, 250)

def resize_with_padding(img, size):
    h, w = img.shape[:2]
    scale = min(size[0] / w, size[1] / h)
    new_w, new_h = int(w * scale), int(h * scale)

    # Redimensionamos manteniendo la relación de aspecto
    interpolation = cv2.INTER_AREA if scale < 1 else cv2.INTER_CUBIC
    resized_img = cv2.resize(img, (new_w, new_h), interpolation=interpolation)

    # Calculamos el color medio del borde en lugar de negro
    border_color = [int(np.mean(resized_img[..., i])) for i in range(3)]

    top_pad = (size[1] - new_h) // 2
    bottom_pad = size[1] - new_h - top_pad
    left_pad = (size[0] - new_w) // 2
    right_pad = size[0] - new_w - left_pad

    # Aplicamos los bordes
    padded_img = cv2.copyMakeBorder(resized_img, top_pad, bottom_pad, left_pad, right_pad, 
                                    cv2.BORDER_CONSTANT, value=border_color)
    return padded_img

# Directorios
input_folder = "../04_Extra/ID/IMAGES"
output_folder = "../04_Extra/ID/IMAGES_PREPROC"
os.makedirs(output_folder, exist_ok=True)

valid_extensions = (".jpg", ".jpeg", ".png", ".bmp")

for set_folder in os.listdir(input_folder):
    set_path = os.path.join(input_folder, set_folder)
    output_set_path = os.path.join(output_folder, set_folder)

    if os.path.isdir(set_path):
        os.makedirs(output_set_path, exist_ok=True)

        for img_name in os.listdir(set_path):
            if not img_name.lower().endswith(valid_extensions):
                continue  # Ignorar archivos no válidos

            img_path = os.path.join(set_path, img_name)
            img = cv2.imread(img_path)

            if img is not None:
                img_padded = resize_with_padding(img, IMAGE_SIZE)
                cv2.imwrite(os.path.join(output_set_path, img_name), img_padded)
            else:
                print(f"⚠️ Advertencia: No se pudo leer {img_path}, omitiendo archivo.")
```

***Modelo de Entrenamiento de CNN optimizado***

```python
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os

# 🟢 Configuración de hiperparámetros
IMAGE_SIZE = (250, 250)
BATCH_SIZE = 32
EPOCHS = 50
LEARNING_RATE = 0.0001  # Aprendizaje más estable
NUM_CLASSES = len(os.listdir("../04_Extra/ID/IMAGES_PREPROC"))  # Contar el número de carpetas (clases)

# 🟢 Generador de imágenes con Data Augmentation
train_datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    rescale=1./255,  # Normalización automática
    validation_split=0.2  # Separar 20% de datos para validación
)

# Cargar imágenes desde la carpeta
IMAGE_DIR = "../04_Extra/ID/IMAGES_PREPROC"

train_generator = train_datagen.flow_from_directory(
    IMAGE_DIR,
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='sparse',
    subset="training"
)

val_generator = train_datagen.flow_from_directory(
    IMAGE_DIR,
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='sparse',
    subset="validation"
)

# 🟢 Construcción de la Red CNN Optimizada
def build_lego_classifier(input_shape=(300, 300, 3), num_classes=NUM_CLASSES):
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=input_shape),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),

        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),

        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),

        layers.Conv2D(256, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),

        layers.GlobalAveragePooling2D(),  # Reemplaza Flatten para evitar sobreajuste
        layers.Dense(512, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')
    ])

    # Compilación del modelo con optimización avanzada
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    return model

# 🟢 Inicializar el modelo
model = build_lego_classifier()

# 🟢 Callbacks para mejorar el entrenamiento
callbacks_list = [
    callbacks.ModelCheckpoint("best_model.h5", monitor='val_accuracy', save_best_only=True, mode='max'),
    callbacks.EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True),
    callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)
]

# 🟢 Entrenamiento del Modelo
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=val_generator,
    callbacks=callbacks_list
)

# 🟢 Guardar el modelo entrenado
model.save("lego_classifier_final.h5")
print("\n✅ Modelo entrenado y guardado correctamente")
```

***Cargamos y usamos el modelo para clasificar nuevas imágenes***

```python
import numpy as np
from tensorflow.keras.models import load_model

# Cargar el modelo entrenado
model = load_model("lego_classifier_final.h5")

def predict_lego_set(image_path):
    """Carga una imagen y predice a qué set de LEGO pertenece."""
    img = load_and_preprocess_image(image_path)
    img = np.expand_dims(img, axis=0)  # Añadir dimensión batch
    
    prediction = model.predict(img)
    predicted_class = np.argmax(prediction)  # Obtener la clase con mayor probabilidad
    
    return predicted_class

# Prueba con una imagen
test_image = "../04_Extra/ID/IMAGEN4.jpg"
predicted_set = predict_lego_set(test_image)
print(f"Set de LEGO Predicho: {predicted_set}")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.regularizers import l2
import os

# 🔹 Configuración de hiperparámetros
IMAGE_SIZE = (300, 300)
BATCH_SIZE = 32
EPOCHS = 50
LEARNING_RATE = 0.0001  
NUM_CLASSES = len(os.listdir("../04_Extra/ID/IMAGES_PREPROC"))  # Contar clases

# 🔹 Generador de imágenes con Data Augmentation mejorado
train_datagen = ImageDataGenerator(
    rotation_range=30,       # Más variabilidad en la rotación
    width_shift_range=0.3,   # Desplazamiento más amplio
    height_shift_range=0.3,  # Desplazamiento más amplio
    shear_range=0.3,         # Más distorsión
    zoom_range=0.3,          # Más zoom
    brightness_range=[0.7, 1.3],  # Cambios de brillo aleatorios
    horizontal_flip=True,    # Volteo horizontal
    fill_mode='nearest',    
    rescale=1./255,         
    validation_split=0.2  # Separar 20% de datos para validación
)

# 🔹 Cargar imágenes desde la carpeta
IMAGE_DIR = "../04_Extra/ID/IMAGES_PREPROC"

train_generator = train_datagen.flow_from_directory(
    IMAGE_DIR,
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='sparse',
    subset="training"
)

val_generator = train_datagen.flow_from_directory(
    IMAGE_DIR,
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='sparse',
    subset="validation"
)

# 🔹 Construcción de la Red CNN Optimizada con Regularización L2
def build_lego_classifier(input_shape=(300, 300, 3), num_classes=NUM_CLASSES):
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', padding='same', 
                      kernel_regularizer=l2(0.001), input_shape=input_shape),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),

        layers.Conv2D(64, (3, 3), activation='relu', padding='same', 
                      kernel_regularizer=l2(0.001)),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),

        layers.Conv2D(128, (3, 3), activation='relu', padding='same', 
                      kernel_regularizer=l2(0.001)),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),

        layers.Conv2D(256, (3, 3), activation='relu', padding='same', 
                      kernel_regularizer=l2(0.001)),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),

        layers.GlobalAveragePooling2D(),
        layers.Dense(512, activation='relu', kernel_regularizer=l2(0.001)),
        layers.Dropout(0.5),  
        layers.Dense(num_classes, activation='softmax')
    ])

    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    return model

# 🔹 Inicializar el modelo
model = build_lego_classifier()

# 🔹 Callbacks para mejorar el entrenamiento
callbacks_list = [
    callbacks.ModelCheckpoint("best_model.h5", monitor='val_accuracy', save_best_only=True, mode='max'),
    callbacks.EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True),
    callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)
]

# 🔹 Entrenamiento del Modelo
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=val_generator,
    callbacks=callbacks_list
)

# 🔹 Guardar el modelo entrenado
model.save("lego_classifier_final.h5")
print("\n✅ Modelo entrenado y guardado correctamente")
```
